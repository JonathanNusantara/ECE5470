#### In this part, we explored on how to create a dataset that is modeled after the MNIST dataset. The goal is to use these images that we have created as test images for a logistic regression model that was trained with the MNIST dataset.

#### Below are the steps in processing the images taken as pictures from a phone to have properties similar to MNIST dataset:
#### The images are first prepared by handwriting a total of 60 drawings. The digits 0 to 9 were first drawn, five each. Then, A to J were drawn to be some symbols that are not 0-9 digits. Pictures of the drawings were taken and uploaded to the computer, which will then be processsed as below. The first step is to import the required libraries. The PIl library is used to load and save images and the skimage.orphology is used for the morphological filtering.

```python
import numpy as np
from PIL import Image
import os
from skimage.morphology import dilation, disk
from sklearn.metrics import confusion_matrix



## A more elegant preserntation for a confusion matrix
import itertools
def plot_confusion_matrix(cm, classes1, classes2,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks1 = np.arange(len(classes1))
    tick_marks2 = np.arange(len(classes2))
    plt.xticks(tick_marks1, classes1, rotation=45)
    plt.yticks(tick_marks2, classes2)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

#### Then, the image is loaded using Image.open() function from PIL. In the following, we are loading 2 images for comparison. The image is first loaded and converted to grayscale. It is loaded as an array, in which we then perform thresholding based on the pixel value. Since we wanted to have a light foreground (pixel value 255) and dark background (pixel value 0), so if the pixel value is smaller than the threshold (black handwritten drawing), then it will have pixel value of 255. The threshold of 80 is chosen after some experimentation to provide the best thresholded output. The two images after thresholding is displayed below.

```python
filename = "img20.jpg"
filename2 = "img40.jpg"

# Convert to binary image by converting to grayscale and threshold
original_image = np.array(Image.open(filename).convert('L'))
print("original image")
display(Image.fromarray(np.uint8(original_image)))
thresh = 80
original_image_array = (original_image < thresh) * 255 # Inverse the value of 255 and 0
print("The original image after thresholding")
display(Image.fromarray(np.uint8(original_image_array)))
print()

#===========================Second Example===========================

# Convert to binary image by converting to grayscale and threshold
original_image2 = np.array(Image.open(filename2).convert('L'))
print("original image")
display(Image.fromarray(np.uint8(original_image2)))
thresh = 80
original_image_array2 = (original_image2 < thresh) * 255 # Inverse the value of 255 and 0
print("The original image after thresholding")
display(Image.fromarray(np.uint8(original_image_array2)))
```

#### Then, a bounding box is created around just the foreground or digit in the image. The list of x and y index that have values bigger than 0 are first extracted using np.where() function. The outputs are two lists containing index values. Then we found the max and min x and y array index and a new array is created based on those arrays, which would create a bounding array of the foreground. This array is then converted into an image using Image.fromarray() function. The results of the bounded digit in the images are as below.

```python
# Bounding box
x,y = np.where(original_image_array) # Get list of x and y values > 0
x_max, x_min, y_max, y_min = x.max(), x.min(), y.max(), y.min()
clipped_image = original_image_array[x_min:x_max,y_min:y_max]
print("Bounded image of digits")
display(Image.fromarray(np.uint8(clipped_image)))
print()

#===========================Second Example===========================

x2,y2 = np.where(original_image_array2) # Get list of x and y values > 0
x_max2, x_min2, y_max2, y_min2 = x2.max(), x2.min(), y2.max(), y2.min()
clipped_image2 = original_image_array2[x_min2:x_max2,y_min2:y_max2]
print("Bounded image of digits")
display(Image.fromarray(np.uint8(clipped_image2)))
```

#### Now that we have have the bounded image, it is time to downsize the bounded image to 20x20 using interpolation while keeping its aspect ratio. One of the reason to downsize is to have a consistent size across all images. The thumbnail() function is used to interpolate and downsize the image. The function is chosen because it allows resizing while keeping the aspect ratio. The bicubic interpolation is chosen because it will provide a good averaging, because it will average with 16 neighboring pixels. Since we are keeping the aspect ratio, it may be the case that either the height or width is not of length of 20 pixels. So, the way to fix this is to create a new image of size 20x20 using Image.new(), then pasting the current image in the middle of the 20x20 image using paste(). In order to perform this, the offset in both the x and y axis needs to be calculated by subtracting the width or height from 20, then divide it by 2 to get the distance from one side of it. Based on the offset, the image is pasted to the 20x20 image ant the result is as below. For visualization purposes, the  image is enlarged again.

```python
# Resize along with aspect ratio and use bicubic interpolation to resize to 20x20
clipped_image = Image.fromarray(np.uint8(clipped_image))
im_thumbnail = clipped_image.thumbnail((20, 20),Image.BICUBIC)
width, height= clipped_image.size[0], clipped_image.size[1]
padded_image = Image.new("L", (20, 20))
offset_x = (20 - width) // 2
offset_y = (20 - height) // 2
padded_image.paste(clipped_image, (offset_x, offset_y))
print("Resized bounded image to 20x20 using interpolation and keeping aspect ratio")
display(padded_image)

print("Above images resized for visualization")
clipped_image_resized = padded_image.resize((200, 200))
display(clipped_image_resized)
print()
#===========================Second Example===========================

clipped_image2 = Image.fromarray(np.uint8(clipped_image2))
im_thumbnail2 = clipped_image2.thumbnail((20, 20),Image.BICUBIC)
width2, height2= clipped_image2.size[0], clipped_image2.size[1]
padded_image2 = Image.new("L", (20, 20))
offset_x2 = (20 - width) // 2
offset_y2 = (20 - height) // 2
padded_image2.paste(clipped_image2, (offset_x2, offset_y2))
print("Resized bounded image to 20x20 using interpolation and keeping aspect ratio")
display(padded_image2)

print("Above images resized for visualization")
clipped_image_resized2 = padded_image2.resize((200, 200))
display(clipped_image_resized2)
```

#### Next, the goal is to have the current 20x20 image to be translated to a 28x28 image and is centered based on the center of mass. This is to have the consistency across all images. In this step, the center of mass is calculated. The center of mass for x and y axis is calculated from the zeroth and first order moments. The COM of x is m10/m00 and y is m01/m00. m00 is the mass, so it is calculated by counting the nonzero pixels in the image. For the m01 and m10, they are calculated using the for loop, in which the values of the index for both axis is summed up when pixel is not 0. We have previously learned that the first order moments are the weighted sum of the pixels based on the distance of the pixel from the origin. The COM of x and y are then calculated. The program is as below.

```python
# Find Center of Mass
m00 = np.count_nonzero(padded_image)
padded_image_array = np.array(padded_image)

m10 = 0 # x
m01 = 0 # y
for i in range(20):
    for j in range(20):
        if padded_image_array[i][j] > 0:
            m10 += j
            m01 += i

com_x, com_y = m10//m00, m01//m00

#===========================Second Example===========================

m002 = np.count_nonzero(padded_image2)
padded_image_array2 = np.array(padded_image2)

m102 = 0 # x
m012 = 0 # y
for i in range(20):
    for j in range(20):
        if padded_image_array2[i][j] > 0:
            m102 += j
            m012 += i

com_x2, com_y2 = m102//m002, m012//m002
```

#### Finally, it is time to paste the 20x20 image that we have to a 28x28 background image and translate it based on the COM value that we have calculated. A new background image of size 28x28 is created. Then, the 20x20 image is pasted on it, with translation of 13 subtracted by the COM on each axis. The digit 13 is the index 27 divided by 2, which results to 13.5, but with the decimal removed. Digit 27 is the maximum index. Now that we have the image array ready, it is converted to an image an is saved in .png format. It is saved to the idata folder directory, which will contain all of the images. The final two images of digit 3 and 7 are as shown below. For visualization purposes, the  image is enlarged again.

```python
# Translate to 28x28 image using Center of Mass
final_image = Image.new("L", (28, 28))
final_image.paste(padded_image, (13-com_y, 13-com_x)) # the middle is index 13.5
name, ext = os.path.splitext(filename)
path = os.path.join(os.getcwd(), "idata/")
name = "img20"
Image.fromarray(np.uint8(final_image)).save(path + name + ".png")
print("Final image")
display(final_image)

print("Resized final image for visualization")
final_image_resized = final_image.resize((200, 200))
display(final_image_resized)
print()

#===========================Second Example===========================

final_image2 = Image.new("L", (28, 28))
final_image2.paste(padded_image2, (13-com_y2, 13-com_x2)) # the middle is index 13.5
name2, ext2 = os.path.splitext(filename2)
name = "img40"
Image.fromarray(np.uint8(final_image)).save(path + name + ".png")
print("Final image")
display(final_image2)

print("Resized final image for visualization")
final_image_resized2 = final_image2.resize((200, 200))
display(final_image_resized2)
```

```python
""" Custom datatset loader
    based on https://github.com/utkuozbulak/pytorch-custom-dataset-examples
"""
import pandas as pd
import imageio

class SimpleDataset():
    def __init__(self, data_path, csv_name, transform = None ):
        """
        Args:
            data_path (string): path to the folder where images and csv files are located
            csv_name (string): name of the csv lablel file
            transform: pytorch transforms for transforms and tensor conversion
        """
        # Set path
        self.data_path = data_path
        # Read the csv file
        self.data_info = pd.read_csv(data_path + csv_name, header=None)
        # First column contains the image paths
        self.image_arr = np.asarray(self.data_info.iloc[:, 0])
        # Second column is the labels
        self.label_arr = np.asarray(self.data_info.loc[:, 1])
        # Calculate len
        self.data_len = len(self.data_info.index)
        
    def __getitem__(self, index):
        # Get image name from the pandas df
        single_image_name = self.image_arr[index]
        # Open image
        img_as_img = imageio.imread(self.data_path + single_image_name)
        
        # Get label(class) of the image based on the cropped pandas column
        single_image_label = self.label_arr[index]
        
        return (img_as_img, single_image_label)

    def __len__(self):
        return self.data_len
```

The training set is standardized so that none of the features can dominate the model, which causes other features to have no impact/contribution to the prediction.
```python
import numpy as np
from PIL import Image
import os
from skimage.morphology import dilation, disk
from sklearn.metrics import confusion_matrix



## A more elegant preserntation for a confusion matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

### Multi-layer perceptron

#### Next, we are exploring the multi-layer perceptron. The multi-layer perceptron is a class of neural network that is made up of multiple layers of perceptron. It will calculate the weights through the iterations of the feedforward and back propagations during the model's training. The multi-layer perceptron will be able to predict data that are not linearly separable due to having multiple-layers. We will be comparing on what will happen when different hidden layers are used. In this test, we are using 1 and 2 hidden layers. The entire data set is used as both the training and testing will be pretty quick. The max iteration is set as 100 to allow the weights to converge and make the best prediction.

```python
# rescale training and testing to be less
first_index = int(60000/1)
second_index = int(60000/1+10000/1)
X_train, X_test = X[:first_index], X[first_index:second_index]
y_train, y_test = y[:first_index], y[first_index:second_index]
```

```python
# Multi-layer perceptron with 1 hidden layer

start_time = time.time()
MLP1 = MLPClassifier(hidden_layer_sizes=(50,), max_iter=100, alpha=1e-4, solver='sgd', verbose=False, tol=1e-4, random_state=1, learning_rate_init=.1)

#Train the classifier
MLP1.fit(X_train,y_train)
train_time_mlp1 = time.time() - start_time
print("Training time %.3f seconds" % train_time_mlp1)

#Evaluate the result
start_time = time.time()
score_mlp1 = MLP1.score(X_test,y_test)
test_time_mlp1 = time.time() - start_time
print("Test time %.3f seconds" % test_time_mlp1)
print("Test score with 1 hidden-layer MLP is: %.4f" % score_mlp1)

#Confusion Matrix
y_predict = MLP1.predict(X_test)
cfm = confusion_matrix(y_test, y_predict)
plot_confusion_matrix(cfm, classes=range(10), title='Confusion matrix for MNIST with 1-hidden layer MLP')
```

The weights converged in 55 iterations.

```python
# Multi-layer perceptron with 2 hidden layer

start_time = time.time()
MLP2 = MLPClassifier(hidden_layer_sizes=(50,50), max_iter=100, alpha=1e-4, solver='sgd', verbose=False, tol=1e-4, random_state=1, learning_rate_init=.1)

#Train the classifier
MLP2.fit(X_train,y_train)
train_time_mlp2 = time.time() - start_time
print("Training time %.3f seconds" % train_time_mlp2)

#Evaluate the result
start_time = time.time()
score_mlp2 = MLP2.score(X_test,y_test)
test_time_mlp2 = time.time() - start_time
print("Test time %.3f seconds" % test_time_mlp2)
print("Test score with 2 hidden-layer MLP is: %.4f" % score_mlp2)

#Confusion Matrix
y_predict = MLP2.predict(X_test)
cfm = confusion_matrix(y_test, y_predict)
plot_confusion_matrix(cfm, classes=range(10), title='Confusion matrix for MNIST with 2-hidden layer MLP')
```

The weights converged in 64 iterations.

```python
table = [["1 hidden-layer MLP",train_time_mlp1,test_time_mlp1,score_mlp1],["2 hidden-layer MLP",train_time_mlp2,test_time_mlp2,score_mlp2]]
print(tabulate(table, headers=["Classifier","Train time (s)", "Test time (s)", "Score"]))
```

#### The perceptron is a parametric model. We can easily identify that by how the train time is much longer than the test time (unlike a non-parametric model like KNN which has the reverse behavior). Train time is much longer because a the best weights would need to be found through iterations until it converges, while the testing step requires only calculations with the weights and figuring out the prediction from there. The train and test time of using 2 hidden layers took around 50% longer compared to using 1 hidden layer. Using 2 hidden layers also took more iterations: 64 vs 55. However, it is able to provide a better test prediction, although not by much (0.9731 vs 0.974). 
#### Comparing the results in using MLP to KNN without PCA pre-processing, we can see that the test accuracy tends to be higher while the train and test time tends to be shorter. However, this is not a fair comparison because because KNN is using less data set to allow the experiment to be within 10 minutes. It may be the case that less training set would translate to less test accuracy, as a result this comparison will not be fair. One thing that we can tell is that multi-layer perceptron is faster then using KNN given that it is using the entire data set. Looking at the confusion matrix for both multi-layer perceptron, they are pretty consistent. It is interesting to see that the most error came from predicting 5 as 3, which is different to using the KNN model.

### SVM

#### Finally, we experimented with Support Vector Machine. The SVM is a type of non-parametric classifier that maps the training set into a space and separate the different labels with a hyperplane that maximizes the distance between the two closest points. The SVM can use a kernel trick to classify a non-linearly separable data. We will be looking into the linear SVM that works only with linearly separable data and Radial Basis Functions SVM that will be able to classify non-linearly separable data. The RBF SVM have a gaussian function and has KNN-like properties. So, it is like a smoothed-version of KNN due to the gaussian function. We will compare the two SVM models' train and test time and their testing accuracy.

```python
# rescale training and testing to be less
first_index = int(60000/2)
second_index = int(60000/2+10000/2)
X_train, X_test = X[:first_index], X[first_index:second_index]
y_train, y_test = y[:first_index], y[first_index:second_index]
```

```python
## Linear Support-Vector Machine

start_time = time.time()
SVC1 = SVC(kernel = 'linear', C = 1)

#Train the classifier
SVC1.fit(X_train,y_train)
train_time_svc1 = time.time() - start_time
print("Training time %.3f seconds" % train_time_svc1)

#Evaluate the result
start_time = time.time()
score_svc1 = SVC1.score(X_test,y_test)
test_time_svc1 = time.time() - start_time
print("Test time %.3f seconds" % test_time_svc1)
print("Test score with Linear SVM is: %.4f" % score_svc1)

#Confusion Matrix
y_predict = SVC1.predict(X_test)
cfm = confusion_matrix(y_test, y_predict)
plot_confusion_matrix(cfm, classes=range(10), title='Confusion matrix for MNIST with Linear SVM')
```

```python
## Radial Basis Functions Support-Vector Machine

start_time = time.time()
SVC2 = SVC(kernel = 'rbf', C = 1, gamma = 0.5) # initially C = 1 gamma = 0.5

#Train the classifier
SVC2.fit(X_train,y_train)
train_time_svc2 = time.time() - start_time
print("Training time %.3f seconds" % train_time_svc2)

#Evaluate the result
start_time = time.time()
score_svc2 = SVC2.score(X_test,y_test)
test_time_svc2 = time.time() - start_time
print("Test time %.3f seconds" % test_time_svc2)
print("Test score with Radial-Basis SVM is: %.4f" % score_svc2)

#Confusion Matrix
y_predict = SVC2.predict(X_test)
cfm = confusion_matrix(y_test, y_predict)
plot_confusion_matrix(cfm, classes=range(10), title='Confusion matrix for MNIST with RBF SVM')
```

Looking at the performance of K-Nearest Neighbor, it is also a pretty good classifier for this application. Out of all classifiers, the SVM seemed to be the poorest. Although linear SVM still has pretty good predictions of 0.926, the RBF SVM is really poor in making predictions for our application, despite an attempt with first performing PCA.
